#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

```python
# TODO

import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]

poblaciones = cinturon_sur['level5'].unique()
normalizados = {}

for poblacion in poblaciones:
    precios = cinturon_sur[cinturon_sur['level5'] == poblacion]['price']
    precios_norm = (precios - precios.min()) / (precios.max() - precios.min())
    normalizados[poblacion] = precios_norm

plt.figure(figsize=(10, 6))

for poblacion, precios_norm in normalizados.items():
    plt.hist(precios_norm, bins=40, alpha=0.5, label=poblacion)

plt.title('Histogramas de precios por población')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.legend(title='Población')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:

from ipyleaflet import Map, Marker, MarkerCluster, Icon
import pandas as pd
from ipyleaflet import basemaps

map = Map(center=(40.4168, -3.7038), zoom=10, min_zoom=1, max_zoom=20, 
          basemap=basemaps.OpenStreetMap.Mapnik)  

ds = pd.read_csv('assets/real_estate.csv', sep=';')


cinturon_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]
pd.set_option('display.max_columns', None) 


population_colors = {
    "Fuenlabrada": "blue",
    "Leganés": "green",
    "Getafe": "red",
    "Alcorcón": "purple"
}

markers = []

for _, row in cinturon_sur.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    population = row['level5']

    color = population_colors.get(population, "gray")  
    
    icon = Icon(color=color, icon_size=(25, 25))  
    
    marker = Marker(location=(latitude, longitude), draggable=False, 
                    title=population, icon=icon)
    
    markers.append(marker)

marker_cluster = MarkerCluster(markers=markers)
map.add_layer(marker_cluster)

map
```
